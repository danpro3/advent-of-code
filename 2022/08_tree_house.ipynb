```python
# part 1
import numpy as np

# lines = open('inputs/input_08_test.txt','r').read().splitlines()
lines = open('inputs/input_08.txt','r').read().splitlines()

X = len(lines)
Y = len(lines[0])

forest = []
for i,line in enumerate(lines):
    forest.append([int(x) for x in list(line)])
forest = np.array(forest)
# print(forest)

# check visibility along 1 direction
def check_visibiliy(row_of_trees):
    visiblelist = 0*row_of_trees
    visiblelist[0] = 1
    tallest = row_of_trees[0]
    for i in range(1,len(row_of_trees)):
        if row_of_trees[i] > tallest:
            visiblelist[i] = 1
            tallest = row_of_trees[i]
    return visiblelist

# look in each direction
visibility = np.zeros((X, Y, 4), dtype='int')

# look north
for i in range(Y):
    row_of_trees = forest[:,i]
    visiblelist = check_visibiliy(row_of_trees)
    visibility[:,i,0] = visiblelist
# print('visibility looking North = ')
# print(visibility[:,:,0])

# look East
for i in range(X):
    row_of_trees = forest[i,::-1]
    visiblelist = check_visibiliy(row_of_trees)
    visibility[i,::-1,1] = visiblelist
# print('visibility looking East = ')
# print(visibility[:,:,1])

# Look South
for i in range(Y):
    row_of_trees = forest[::-1,i]
    visiblelist = check_visibiliy(row_of_trees)
    visibility[::-1,i,2] = visiblelist
# print('visibility looking South = ')
# print(visibility[:,:,2])

# Look West
for i in range(X):
    row_of_trees = forest[i,:]
    visiblelist = check_visibiliy(row_of_trees)
    visibility[i,:,3] = visiblelist
# print('visibility looking West = ')
# print(visibility[:,:,3])

visible_trees = np.zeros((X, Y), dtype='int')
for x in range(X):
    for y in range(Y):
        if any(visibility[x,y,:] == 1):
            visible_trees[x,y] = 1

print(f'sum of visible trees = {visible_trees.sum()}')

```

```python
# part 2 (continued)
scenic_score = np.zeros((X,Y,4),dtype='int')

# print(forest)
for x in range(X):
    for y in range(Y):
        height = forest[x,y]

        # check north
        if x > 0 and x < X:  # not on the edge
            visible = 1
            i = x-1
            while i>0 and forest[i,y] < height:
                visible += 1
                i -= 1
            scenic_score[x,y,0] = visible

        # check east
        if y >= 0 and y < Y-1:  # not on the edge
            visible = 1
            i = y+1
            while i<Y-1 and forest[x,i] < height:
                visible += 1
                i += 1
            scenic_score[x,y,1] = visible

        # check south
        if x >= 0 and x < X-1:  # not on the edge
            visible = 1
            i = x+1
            while i<X-1 and forest[i,y] < height:
                visible += 1
                i += 1
            scenic_score[x,y,2] = visible

        # check west
        if y > 0 and y < Y:  # not on the edge
            visible = 1
            i = y-1
            while i>0 and forest[x,i] < height:
                visible += 1
                i -= 1
            scenic_score[x,y,3] = visible

# print(scenic_score)

# tally up the multiplication of the last dimension
total_scenic_score = np.prod(scenic_score,axis=2)

# print(total_scenic_score)
print(f'highest total scenic score = {np.max(total_scenic_score)}')
        
```
