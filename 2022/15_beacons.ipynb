```python
# part 1
import re

def merge_ranges(ranges):
    """Merges a set of ranges into a list of non-overlapping ranges."""
    # Sort the ranges by their start values
    ranges.sort(key=lambda x: x[0])
    merged_ranges = []
    current_range = ranges[0]
    for i in range(1, len(ranges)):
        next_range = ranges[i]
        # Check for overlap
        if current_range[1] >= next_range[0]:
            # Merge overlapping ranges
            current_range = (current_range[0], max(current_range[1], next_range[1]))
        else:
            # Add the current range to the merged list
            merged_ranges.append(current_range)
            current_range = next_range
    # Add the last range
    merged_ranges.append(current_range)
    return merged_ranges

# lines = open('inputs/input_15_test.txt','r').read().splitlines()
lines = open('inputs/input_15.txt','r').read().splitlines()
# YY = 10
YY = 2000000 # row of interest
ranges = []
for line in lines:
    S = [int(x) for x in re.findall(r'-?\d+', line)]  # sensor info [sensor:x,y  nearest beacon:x,y]
    # print(S)
    S_range = abs(S[0]-S[2]) + abs(S[3]-S[1])
    reach = S_range - abs(S[1] - YY)
    if reach >= 0: # need to block off some positions
        newrange = [S[0]-reach, S[0]+reach]
        print(f'sensor: {S[0],S[1]}, range: {S_range}, can reach: {reach}, newrange = {newrange}')
        ranges.append(newrange)
        ranges = merge_ranges(ranges)

# sum the ranges:
print(f'ranges = {ranges}')
sum_ranges = 0
for ranges1 in ranges:
    sum_ranges += ranges1[1] - ranges1[0] + 1

sum_ranges -= 1 # subract 1 for the beacon in the range
print(f'sum of ranges = {sum_ranges}')
```

```python
# part 2

# parse the input first
SS = []
for line in lines:
    S = [int(x) for x in re.findall(r'-?\d+', line)]  # sensor info [sensor:x,y  nearest beacon:x,y]
    SS.append(S)

# YYmax = 20 # max row of interest
YYmax = 4000000 # max row of interest
YY = 1000000
ranges = []
while YY < YYmax and len(ranges) <= 1 :
    ranges = []
    for i in range(len(lines)):
        S = SS[i] # sensor info [sensor:x,y  nearest beacon:x,y]
        # print(S)
        S_range = abs(S[0]-S[2]) + abs(S[3]-S[1])
        reach = S_range - abs(S[1] - YY)
        if reach >= 0: # need to block off some positions
            newrange = [max(S[0]-reach, 0), min(S[0]+reach, YYmax)]
            # print(f'sensor: {S[0],S[1]}, range: {S_range}, can reach: {reach}, newrange = {newrange}')
            ranges.append(newrange)
            ranges = merge_ranges(ranges)
    if YY % 100000 == 0:
        print(f'YY = {YY}, ranges = {ranges}, len(ranges) = {len(ranges)}')
    YY += 1

print(f'YY = {YY}, ranges = {ranges}, len(ranges) = {len(ranges)}')
x, y = ranges[0][1]+1, YY-1
print(f'coord of distress beacon (x,y)= ({x, y})')
print(f'tuning frequency = {4000000*x + y}')
```
