```python
# Part 1
lines = [x.split() for x in open("input_07.txt").read().splitlines()]
hands = [[x[0],int(x[1])] for x in lines]

deck = {'A' : 14, 'K' : 13, 'Q' : 12, 'J' : 11, 'T' : 10,  '9' : 9, '8' : 8, '7' : 7, '6' : 6, '5' : 5, '4' : 4, '3' : 3, '2' : 2}
poker_ranks = {'5 kind': 7, '4 kind': 6, 'full house': 5, '3 kind': 4, '2 pair': 3, '1 pair': 2, 'high card': 1}
print(hands)

max_tiebreak_value = 0
# find the ranks
for h,hand in enumerate(hands):
   
   # find what the hand is
   cards = [hand[0].count(x) for x in deck.keys()]
   if cards.count(5) == 1:
      hand.append(poker_ranks.get('5 kind'))
      # print(h,cards,'5 kind')
   elif cards.count(4) == 1:
      hand.append(poker_ranks.get('4 kind'))
      # print(cards,'4 kind')
   elif cards.count(3) == 1 and cards.count(2) == 1:
      hand.append(poker_ranks.get('full house'))
      # print(cards,'full house')
   elif cards.count(3) == 1 and cards.count(2) == 0:
      hand.append(poker_ranks.get('3 kind'))
      # print(cards,'3 kind')
   elif cards.count(2) == 2:
      hand.append(poker_ranks.get('2 pair'))
      # print(cards,'2 pair')
   elif cards.count(3) == 0 and cards.count(2) == 1:
      hand.append(poker_ranks.get('1 pair'))
      # print(cards,'1 pair')
   else:
      hand.append(poker_ranks.get('high card'))
      # print(cards,'high card')

   # find a tie_breaker value for each hand
   tiebreak_value = 0
   for c in range(5):
      tiebreak_value += deck.get(hand[0][c])*14**(5-c)
   hand.append(tiebreak_value)
   if hand[3] > max_tiebreak_value:
         max_tiebreak_value = hand[3]

print(hands)
print(max_tiebreak_value)

# find a global value for the hand
for hand in hands:
   hand[2] = hand[2]*10*max_tiebreak_value + hand[3]

# sort
hands = sorted(hands, key=lambda column: column[2])
print(hands)

# find winnings
total_winnings = 0
for i,hand in enumerate(hands):
   total_winnings += (i+1)*hand[1]
print('total winnings =',total_winnings)
```
