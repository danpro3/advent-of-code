```python
# part 1
def print_grid(G):
    temp = [''.join(x) for x in G]
    [print(x) for x in temp]


def move(todo):
    steps, r, c, dir, path = todo.pop(0)
    # print(steps,r,c)
    # if we've been here before, end this thread
    if (steps,r,c,dir) in cache:
        # print(r,c,dir,'in cache')
        return todo
    else:
        # add this state to the cache
        # cache[state] = 0
        cache[(steps,r,c,dir)] = 0

    if r == X-1:
        print('finished. r,c =',r,c,'steps =',steps)
        all_hikes.append(steps)
        path.pop(0)
        all_paths.append(path)        
        return todo
    # check up
    if dir != 'S' and r > 0 and grid[r-1][c] in '.^':
        if (steps+1, r-1, c) not in cache:
            path2 = path[:]
            path2.append((r-1,c))
            todo.insert(0,(steps+1, r-1, c, 'N', path2))
    # check right
    if dir != 'W' and c < X-1 and grid[r][c+1] in '.>':
        if (steps+1, r, c+1) not in cache:
            path2 = path[:]
            path2.append((r,c+1))
            todo.insert(0,(steps+1, r, c+1, 'E', path2))
    # check down
    if dir != 'N' and r < Y-1 and grid[r+1][c] in '.v':
        if (steps+1, r+1, c) not in cache:
            path2 = path[:]
            path2.append((r+1,c))
            todo.insert(0,(steps+1, r+1, c, 'S', path2))
    # check left
    if dir != 'E' and c > 0 and grid[r][c-1] in '.<':
        if (steps+1, r, c-1) not in cache:
            path2 = path[:]
            path2.append((r,c-1))
            todo.insert(0,(steps+1, r, c-1, 'W', path2))
    return todo


# main
lines = open('input_23.txt').read().splitlines()
# [print(x) for x in lines]
grid = [list(x) for x in lines]
print(len(grid), len(grid[0]))

X = len(grid)
Y = len(grid[0])
# cache_grid = [['.']*X for x in range(Y)]
# print_cache(cache_grid)

# find start location
startcol = lines[0].find('.')
r,c = (0,startcol)
print('starting block =',r,c)

todo = [(0,r,c,'S',[(r,c)])] # this is the to do list
cache = {}   # list of all places been: (row, col, dir, total steps, value = path list)
all_hikes = []
all_paths = []
i = 0
while len(todo) > 0: # and i < 1000:
    # break
    i += 1
    if i % 1 == 400000:
        print('i =',i,'todo[0] = ',todo[0])
    todo = move(todo)
    # print_cache(cache_grid)
# print_cache(cache_grid)
# print(cache.keys())
print('number of hikes =',len(all_hikes),'longest hike =',max(all_hikes))

idx = all_hikes.index(max(all_hikes))
path_grid = grid[:]
for coords in all_paths[idx]:
    path_grid[coords[0]][coords[1]] = 'O'

# print_grid(path_grid)

```
