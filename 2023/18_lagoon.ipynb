```python
# part 1
def print_grid(grid):
    temp = [''.join(x) for x in grid]
    [print(x) for x in temp]

def dig(path, dir, dist, nextdir):
    r, c, _, _ = path[-1]
    for i in range(dist):
        # (r, c, value, prevdir) = path[-1]
        r -= 1 if dir == 'U' else 0
        r += 1 if dir == 'D' else 0
        c -= 1 if dir == 'L' else 0
        c += 1 if dir == 'R' else 0
        value = 1 if dir in 'UD' else 0  # straights
        # corners
        if i == dist-1:
            # print(f'corner i={i}, r = {r}, c = {c}, dir ={dir}, nextdir = {nextdir}')
            if dir == 'L' and nextdir == 'U':
                value = -.5
            if dir == 'R' and nextdir == 'U': 
                value = .5
            if dir == 'R' and nextdir == 'D':
                value = -.5
            if dir == 'L' and nextdir == 'D':
                value = .5
            if dir == 'U' and nextdir == 'L':
                value = -.5
            if dir == 'U' and nextdir == 'R':
                value = .5
            if dir == 'D' and nextdir == 'L':
                value = .5
            if dir == 'D' and nextdir == 'R':
                value = -.5
            # value = 8
            # print('value =',value)
        path.append((r, c, value, dir))
        # print(path)
        # print(' ')

# main
# lines = open("input_18_test.txt").read().splitlines()
lines = open("input_18.txt").read().splitlines()
path = [(0, 0, 0, 'X')]  # add coords and a value
total_dist = 0
for i,line in enumerate(lines):
    if i < len(lines)-1:
        nextdir, _, _, = lines[i+1].split()
    else:
        nextdir, _, _, = lines[0].split()

    dir, dist, color = line.split()
    dist = int(dist)
    # print(dir, dist, color)
    dig(path, dir, dist, nextdir)
    total_dist += dist

# determine the value of start tile
print('start and end of path =',path[0], path[1])
dir_start, _, _ = lines[0].split()
dir_end, _, _ = lines[-1].split()
print('start and end of path dirs=',dir_start, dir_end)
value = .5
if dir_start in 'UD' and dir_end in 'UD':
    value = 1
if dir_start in 'LR' and dir_end in 'LR':
    value = 0
path[0] = (path[0][0], path[0][1], value, path[1][3])
print('new start and end of path =',path[0], path[1])

path.pop(-1)

# print(path)
print('length path = ',len(path))
print('total distance =', total_dist)

# build a grid
# find minimum of the coords and offset to all positives
X = 0; Y = 0
for i in range(len(path)):
    X = min(X, path[i][0])
    Y = min(Y, path[i][1])
# re-normalize the path
print('min X,Y =',X,Y)
# print('raw path =', path)
for i in range(len(path)):
    path[i] = (path[i][0] - X, path[i][1] -Y, path[i][2], path[i][3])
print('shifted path =', path)
X = 0; Y = 0
for i in range(len(path)):
    X = max(X, path[i][0])
    Y = max(Y, path[i][1])
X += 1; Y += 1
print('size =',X,Y)

grid = [['.']*Y for x in range(X)]
grid_values = [[0]*Y for x in range(X)]
# print(path)
for i in range(len(path)):
    # print(i, path[i][0], path[i][1])
    grid[path[i][0]][path[i][1]] = '#' 
    grid_values[path[i][0]][path[i][1]] = path[i][2]
# print_grid(grid)
# [print(i) for i in grid_values]

# compute the area
tiles_inside = 0
for a in range(1,X):
    for b in range(1,Y):
        if grid[a][b] == '.':
            value = sum(grid_values[a][0:b])
            if value % 2 == 1:
                tiles_inside += 1
    # print(a, tiles_inside)        
print('Number of inside tiles =',tiles_inside)
print('total area =',tiles_inside + len(path))

# 47982 is too high
```
