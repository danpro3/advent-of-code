```python
# part 2
def print_grid(grid):
    temp = [''.join(x) for x in grid]
    [print(x) for x in temp]

def dig(path, dir, dist, nextdir):
    r, c, _, _ = path[-1]
    r -= dist if dir == 'U' else 0
    r += dist if dir == 'D' else 0
    c -= dist if dir == 'L' else 0
    c += dist if dir == 'R' else 0
    value = 1
    path.append((r, c, value, dir))

def shoestring(path):
    area = 0
    for i in range(len(path)-1):
        area += path[i][1]*path[i+1][0] - path[i+1][1]*path[i][0]
    area += path[-1][1]*path[0][0] - path[0][1]*path[-1][0]
    area = 1/2 * abs(area)
    return area

# main
# lines = open("input_18_test.txt").read().splitlines()
lines = open("input_18.txt").read().splitlines()
path = [(0, 0, 0, 'X')]  # add coords and a value
total_dist = 0
for i,line in enumerate(lines):
    # dir, dist, color = line.split() # part 1
    _, _, color = line.split() # part 1
    mapping = {i:d for i,d in enumerate('RDLU')}
    dir = mapping[int(color[-2])]
    dist = int('0x'+color[2:-2],0)
    # print(dir, dist)
    # dist = int(dist)
    # print(dir, dist, color)
    dig(path, dir, dist, dir)
    total_dist += dist
path.pop(0)

# print(path)
print('num verticies = ',len(path))
print('total distance =', total_dist)

# build a grid
# find minimum of the coords and offset to all positives
X = 0; Y = 0
for i in range(len(path)):
    X = min(X, path[i][0])
    Y = min(Y, path[i][1])
# re-normalize the path
print('min X,Y =',X,Y)
# print('raw path =', path)
for i in range(len(path)):
    path[i] = (path[i][0] - X, path[i][1] -Y, path[i][2], path[i][3])
# print('shifted path =', path)
X = 0; Y = 0
for i in range(len(path)):
    X = max(X, path[i][0])
    Y = max(Y, path[i][1])
X += 1; Y += 1
print('size =',X,Y)

# grid = [['.']*Y for x in range(X)]
# grid_values = [[0]*Y for x in range(X)]
# print(path)
# for i in range(len(path)):
#     # print(i, path[i][0], path[i][1])
#     grid[path[i][0]][path[i][1]] = '#' 
#     grid_values[path[i][0]][path[i][1]] = path[i][2]
# print_grid(grid)
# [print(i) for i in grid_values]

# compute the area
print('path length = ',total_dist)
area = shoestring(path)
# shoestring area = enclosed + (1/2(path) - 1)
# adding 1/2(path) is too much, need to knock off the 1 (that last 1 is because 4 * 1/4 corners only add to 1 and not 2)
# Total lagoon = enclosed + path = shoestring + path + 1
print('shoestring area =',area)
print('total area =',area + total_dist/2 + 1)
```
