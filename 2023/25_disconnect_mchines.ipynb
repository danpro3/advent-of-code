```python
# part 1 second half
# find out how many paths to each node
import random as rand
import math
from itertools import islice
import copy

def nth_key(dct, n):
    it = iter(dct)
    # Consume n elements.
    next(islice(it, n, n), None) 
    # Return the value at the current position.
    # This raises StopIteration if n is beyond the limits.
    # Use next(it, None) to suppress that exception.
    return next(it)

def contract_edge(A1,A2):
    BB1 = [x[0] for x in nodes[A1]]
    index = BB1.index(A2)  # find A2 in A1's list
    nodes[A1][0][1] += nodes[A2][0][1] # consolidate the shared edges
    nodes[A1].pop(index)  # removed A2 from A1's list
    BB1.pop(index)
    for B2 in nodes[A2]:
        if B2[0] == A1 or B2[0] == A2:
            continue
        if B2[0] in BB1:
            index = BB1.index(B2[0])
            nodes[A1][index][1] += B2[1]
        else:
            nodes[A1].append(B2)
    del nodes[A2]
    # now need to loop thru the other nodes and consolidate
    1
    for A in nodes:
        BB = [x[0] for x in nodes[A]]
        if A2 in BB:
            index = BB.index(A2)
            nodes[A][index][0] = A1
            BB[index] = A1
        index_A1 = [i for i in range(len(BB)) if BB[i]==A1]
        if len(index_A1) == 2:
            nodes[A][index_A1[0]][1] += nodes[A][index_A1[1]][1]
            nodes[A].pop(index_A1[1])  # removed duplicate A1

def add_ones(nodes):
    for A in nodes:
        BB = nodes[A]
        nodes[A] = []
        for B in BB:
            nodes[A].append([B,1])
        nodes[A].insert(0,[A,1])

# main
line_nodes = build_line_nodes(lines)
nodes = build_nodes(nodes)
add_ones(nodes)
nodes_orig = copy.deepcopy(nodes)
print(f'\nnodes:')
[print(x,nodes[x]) for x in nodes]

# for k in range(5):
#     x = math.floor(len(nodes)*rand.random())
#     print(x, nth_key(nodes,x))

# [print(x,nodes[x]) for x in nodes]

# A = 'jqt'; B = 'rhn'
# print('contact:',A,B)
# contract_edge(A,B)
# print(''); print('nodes:')
# [print(x,nodes[x]) for x in nodes]

run = 0
edge_count = 0
while edge_count != 3:
    run += 1
    nodes = copy.deepcopy(nodes_orig)
    while len(nodes) > 2:
        n = math.floor(len(nodes)*rand.random())
        A = nth_key(nodes,n)
        n = math.floor((len(nodes[A])-1)*rand.random())+1
        B = nodes[A][n][0]
        # print('contact:',A,B)
        contract_edge(A,B)
    for A in nodes:
        edge_count = nodes[A][1][1]
    print(f'run #{run}, edge_count = {edge_count}')

node_count = [nodes[x][0][1] for x in nodes]
print(f'\n{sum(node_count)} nodes:')
[print(x,nodes[x]) for x in nodes]
mult_groups = 1
for A in nodes:
    mult_groups *= nodes[A][0][1]
print(f'mult_groups = {mult_groups}')


```
