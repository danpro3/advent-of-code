```python
# part 1
from itertools import combinations
# c = list(combinations(range(5),2))

def make_justlists(A,BB):
    justlists.append([A])
    for b in BB:
        justlists[-1].append(b)
    return

def consolidate(justlists):
    changed = True
    while changed == True:
        combos = list(combinations(range(len(justlists)),2))
        changed = False
        for combo in combos:
            newlist = set(justlists[combo[0]]) & set(justlists[combo[1]])
            # print(combo, newlist, newlist == set())
            if newlist != set() and justlists[combo[0]] != set() and justlists[combo[1]] != set():
                changed = True
                justlists[combo[0]] = list(set(justlists[combo[0]]) | set(justlists[combo[1]]))
                justlists[combo[1]] = set()
        for i in range(len(justlists)-1,0,-1):
            justlists.remove(justlists[i]) if justlists[i] == set() else 0
    # [print(x) for x in justlists]
    return justlists


def add_new_to_allcontacts(A,BB):  # BB is the destination list
    if A in allcontacts:
        [allcontacts[A].append(B) for B in BB]
    else:
        allcontacts[A] = BB
    for B in BB:
        if B in allcontacts:
            allcontacts[B].append(A)
        else:
            allcontacts[B] = [A]
    return allcontacts


# main
lines = open('inputs/input_25_test.txt').read().splitlines()
# print(lines)
contacts = {}
for line in lines:
    AA = line.split(':')
    BB = AA[1][1:].split(' ')
    # print(AA[0],BB)
    contacts[AA[0]] = BB

allcontacts = {}
justlists = []
for A in contacts:
    add_new_to_allcontacts(A,contacts[A])
    make_justlists(A,contacts[A])

print('contacts:')
[print(x,contacts[x]) for x in contacts]

# print('justlists:')
# [print(x) for x in justlists]

print('allcontacts:')
[print(x,allcontacts[x]) for x in allcontacts]

# justlists = consolidate(justlists)
# print('justlists')
# [print(x) for x in justlists]


# sample: remove the pairs: hfx/pzl, bvb/cmg, nvd/jqt
# contacts['pzl'].remove('hfx')
# contacts['cmg'].remove('bvb')
# contacts['jqt'].remove('nvd')

# other:
# contacts['jqt'].remove('rhn')
# contacts['lsr'].remove('lhk')
# contacts['cmg'].remove('qnr')


allcontacts = {}
justlists = []
for A in contacts:
    add_new_to_allcontacts(A,contacts[A])
    make_justlists(A,contacts[A])
print('allcontacts')
[print(x) for x in allcontacts]

justlists = consolidate(justlists)
print('justlists')
[print(x) for x in justlists]

```
