```python
import re
import math

sum_part_numbers = 0
sum_gears = 0
gears = []

with open("input_03.txt", "r") as file:
   filestr = file.read()
# print(filestr)
numcols = filestr.find('\n') + 1  # number of columns in the engine schematic
filestr = filestr.replace('\n','.')
numrows = len(filestr) // numcols
print('numrows =',numrows,', numcols = ',numcols)
number_list = re.findall('\d+',filestr)
# print(number_list)
number_idx = [(m.start(0), m.end(0)-1) for m in re.finditer('\d+', filestr)]
print(number_idx)
for i,num in enumerate(number_list):
   valid_number = False
   col_start = number_idx[i][0] % numcols
   col_end = number_idx[i][1] % numcols
   row = number_idx[i][0] // numcols
   # print('row = ',row,', column from',col_start,'to',col_end)
   
   # check same row
   if col_start > 0:
      # print(number_list[i], filestr[number_idx[i][0]-1])
      if re.match('[^.]',filestr[number_idx[i][0]-1]):
         valid_number = True
         symbols = filestr[number_idx[i][0]-1]
        #  print(num,'validated to the left with',symbols)
         if symbols == '*':
            gears.append((number_idx[i][0]-1,int(num)))
   if col_end < numcols-1:
      # print(number_list[i], filestr[number_idx[i][1]+1])
      if re.match('[^.]',filestr[number_idx[i][1]+1]):
         valid_number = True
         symbols = filestr[number_idx[i][1]+1]
        #  print(num,'validated to the right with',symbols)
         if symbols == '*':
            gears.append((number_idx[i][1]+1,int(num)))

   # check above
   if row > 0:
      # print(number_list[i],'above = ',filestr[number_idx[i][0]-numcols:number_idx[i][1]-numcols+1])
      if re.findall('[^.]',filestr[number_idx[i][0]-numcols:number_idx[i][1]-numcols+1]):
         valid_number = True
         symbols = filestr[number_idx[i][0]-numcols:number_idx[i][1]-numcols+1]
        #  print(num,'validated above with',symbols)
         gears_ind = symbols.find('*')
         if gears_ind > -1:
            gears.append((number_idx[i][0]-numcols+gears_ind,int(num)))

      if col_start > 0:  # check upper left
         # print(number_list[i],'upper left corner = ',filestr[number_idx[i][0]-numcols-1],'index =',number_idx[i][0]-numcols-1)
         if re.match('[^.]',filestr[number_idx[i][0]-numcols-1]):
            valid_number = True
            symbols = filestr[number_idx[i][0]-numcols-1]
            # print(num,'validated to the upper left with',symbols)
            if symbols == '*':
               gears.append((number_idx[i][0]-numcols-1,int(num)))

      if col_end < numcols-1: # check upper right
         if re.match('[^.]',filestr[number_idx[i][1]-numcols+1]):
            valid_number = True
            symbols = filestr[number_idx[i][1]-numcols+1]
            # print(num,'validated to the upper right with',symbols)
            if symbols == '*':
               gears.append((number_idx[i][1]-numcols+1,int(num)))

   # check below
   if row < numrows - 1:
      if re.findall('[^.]',filestr[number_idx[i][0]+numcols:number_idx[i][1]+numcols+1]):
         valid_number = True
         symbols = filestr[number_idx[i][0]+numcols:number_idx[i][1]+numcols+1]
        #  print(num,'validated below with',symbols)
         gears_ind = symbols.find('*')
         if gears_ind > -1:
            gears.append((number_idx[i][0]+numcols+gears_ind,int(num)))

      if col_start > 0:  # check lower left
         if re.match('[^.]',filestr[number_idx[i][0]+numcols-1]):
            valid_number = True
            symbols = filestr[number_idx[i][0]+numcols-1]
            # print(num,'validated to the lower left with',symbols)
            if symbols == '*':
               gears.append((number_idx[i][0]+numcols-1,int(num)))

      if col_end < numcols-1: # check lower right
         if re.match('[^.]',filestr[number_idx[i][1]+numcols+1]):
            valid_number = True
            symbols = filestr[number_idx[i][1]+numcols+1]
            # print(num,'validated to the lower right with',symbols)
            if symbols == '*':
               gears.append((number_idx[i][1]+numcols+1,int(num)))

   if valid_number:
      sum_part_numbers = sum_part_numbers + int(num)         

print('Sum of part numbers = ',sum_part_numbers)
print(gears)
for k in range(len(gears)-1):
   for m in range(k+1,len(gears)):
      if gears[k][0] == gears[m][0]:
         sum_gears += gears[k][1] * gears[m][1]
print('Sum of gear ratios =',sum_gears)


```
