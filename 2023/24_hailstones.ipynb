# Part 1 
This a system of linear equations solver
$$ \frac{y-y_0}{x-x_0} = m \textnormal{, where } m = \frac{v_y}{v_x} $$
$$ mx - mx_0 = y-y_0 $$
$$ mx - y = mx_0-y_0 $$
$$ ax + by = c \\ 
\textnormal{where } a = m, b = -1, c = mx_0-y_0 $$
Need to solve: $Ax = B$ for each pair. <br>
$$\begin{pmatrix}
a_1&b_1\\a_1&b_1
\end{pmatrix}\cdot
\begin{pmatrix}x\\y\end{pmatrix}=\begin{pmatrix}c_1\\c_2\end{pmatrix}$$

$$\begin{pmatrix}
v_{y1}/v_{x1}&-1\\v_{y2}/v_{x2}&-1
\end{pmatrix}\cdot
\begin{pmatrix}x\\y\end{pmatrix}=\begin{pmatrix}x_1\cdot v_{y1}/v_{x1}-y_1\\x_2\cdot v_{y2}/v_{x2}-y_2\end{pmatrix}$$

Then determine if the intersection is in the past for future
Time to read any point $x_1,y_1$ is
$$ x_1 = x_0 + v_x t \; \rightarrow \hspace{3mm} t = \frac{x_1-x_0}{v_x}$$

```python
# part 1
from itertools import combinations
import numpy as np
# c = list(combinations(range(5),2))
# print(c)
# a = np.array([[1, 2], [3, 5]])
# b = np.array([1, 2])
# x = np.linalg.solve(a, b)

def get_intersection(combo):
    X1 = pos[combo[0]]
    V1 = vels[combo[0]]
    X2 = pos[combo[1]]
    V2 = vels[combo[1]]
    # matrix solver for Ax = B
    # A = np.array([[1, 2], [3, 5]])
    # B = np.array([1, 2])
    # x = np.linalg.solve(A, B)
    A = np.array([[V1[1]/V1[0], -1], [V2[1]/V2[0], -1]])
    B = np.array([V1[1]/V1[0]*X1[0]-X1[1], V2[1]/V2[0]*X2[0]-X2[1]])
    X = np.linalg.solve(A, B)
    T = [0,0]
    T[0] = (X[0] - X1[0])/V1[0]
    T[1] = (X[0] - X2[0])/V2[0]
    return X,T

# main
# lines = open('inputs/input_24_test.txt').read().splitlines(); bounds = (7,27)  # for test input
lines = open('inputs/input_24.txt').read().splitlines(); bounds = (200000000000000,400000000000000)  # for real input

# [print(x) for x in lines]
pos = []
vels = []
for line in lines:
    hailstones = line.split('@')
    pos.append([int(x) for x in hailstones[0].split(',')])
    vels.append([int(x) for x in hailstones[1].split(',')])
# print(f'pos = {pos}')
# print(f'vels = {vels}')
print(f'Total number of hailstones = {len(pos)}')

total_collisions = 0
for combo in combinations(range(len(pos)),2):
    # check for parallel
    m1y = vels[combo[0]][1]/vels[combo[0]][0]
    m1z = vels[combo[0]][2]/vels[combo[0]][0]
    m2y = vels[combo[1]][1]/vels[combo[1]][0]
    m2z = vels[combo[1]][2]/vels[combo[1]][0]
    if m1y == m2y and m1z == m2z:
        1 # parallel
        print(f'combo {combo} are parallel')
        print('hail #',combo[0],':',pos[combo[0]],vels[combo[0]])
        print('hail #',combo[1],':',pos[combo[1]],vels[combo[1]])
    else:
        if m1y != m2y:
            X,T = get_intersection(combo)
            # print(f'combo {combo} intersect at X = ({X[0]:.2f},{X[1]:.2f}) and T = {T[0]:.2f},{T[1]:.2f}')
            if bounds[0] <= X[0] <= bounds[1] \
            and bounds[0] <= X[1] <= bounds[1] \
            and all([t > 0 for t in T]):
                total_collisions += 1

print(f'total collisions = {total_collisions}')

```
