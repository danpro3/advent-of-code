```python
# Part 1

with open("input_05.txt", "r") as file:
   filestr = file.read()
filestr = filestr.replace('\n',' ')
lines = filestr.split(':')
# print('lines =',lines)
seeds = [int(x) for x in lines[1].split()[:-2]]
# print('seeds = ',seeds)
codes = []
for i in range(2,len(lines)):
   code = [int(x) for x in lines[i].split()[:-2]]
   codes.append(code)
codes[-1] = [int(x) for x in lines[-1].split()]
# print('codes =',codes)
locations = []
for seed in seeds:
   # print('seed =',seed)
   dest = seed
   for code in codes:
      # print('code =',code)
      source = dest
      groups = len(code)//3
      for group in range(groups):  # go search for ranged codes
         if source in range(code[3*group+1], code[3*group+1] +  code[3*group+2]):
            dest = code[3*group+0] + source - code[3*group+1]
   # print('seed =',seed,'source =',source,', dest =',dest)
   locations.append(dest)
# print('locations =',locations)
print('min location =',min(locations))

```

```python
# Part 2

def range_splitter(i,A,B,group):
   # 1: in the group range
   need_to_recheck = False
   low = max(A[i][0],group[1])
   high = min(A[i][0]+A[i][1]-1,group[1]+group[2]-1)
   if high-low+1 > 0:
      # convert it to next level
      # print('in group: ',(low,high-low+1))
      B.append((low+group[0]-group[1],high-low+1))
      # print('A =',A,'B =',B)
      # need_to_recheck = True

   jump = 0
   # 2: below the group range
   low = A[i][0]
   high = min(A[i][0]+A[i][1]-1,group[1]-1)
   if low < group[1]:
      # print('below')
      A.insert(1,(low,high-low+1))
      jump += 1
      
   # 3: above the group range
   low = max(A[i][0],group[1]+group[2])
   high = A[i][0]+A[i][1]-1
   if high > group[1]+group[2]-1:
      # print('above')
      A.insert(1,(low,high-low+1))
      jump += 1

   # Done checking A[i], remove the additions
   A.remove(A[i])
   return A,B,jump

# main
with open("input_05.txt", "r") as file:
   filestr = file.read()
filestr = filestr.replace('\n',' ')
lines = filestr.split(':')
# print('lines =',lines)
seeds = ([int(x) for x in lines[1].split()[:-2]])
# print('seeds =',seeds)
seed_ranges = [(seeds[i],seeds[i+1]) for i in range(0,len(seeds),2)]
# print('seed_ranges =',seed_ranges)

maps = []
for i in range(2,len(lines)):
   if i < len(lines)-1:
      mapp = [int(x) for x in lines[i].split()[:-2]]
   else:
      mapp = [int(x) for x in lines[i].split()]
   # print(mapp)
   group = [(mapp[i],mapp[i+1],mapp[i+2]) for i in range(0,len(mapp),3)];
   maps.append(group)   
# print('maps =',maps)

locations = []
# each mapp takes A -> B
for s,seed_range in enumerate(seed_ranges):
   A = [seed_range]
   for m,mapp in enumerate(maps):
      # print('mapp =',mapp)
      B = []
      for group in mapp:
         # print('group =',group)
         need_to_recheck = False
         i = 0
         while i < len(A):
            # print('Start of A loop. A =',A,'index =',i)
            A,B,jump = range_splitter(i,A,B,group)
            # print('Splitter: A =',A,'B =',B)
            i += jump
            # print('End of A loop. A =',A,'i =',i)
         # print('end of group')

      # after each mapp, append list B onto the whatever is left in A
      for Bi in B:
         A.append(Bi)
      # print('>>> End of mapp #',m,'in seed range',s,'New A =',A)
   
   for i,L in enumerate(A):
      if i == 0:
         min_locations = A[i][0]
      else:
         if A[i][0] < min_locations:
            min_locations = A[i][0]
   locations.append(min_locations)
   # print('locations =',locations)
print('min location =',min(locations))

```
